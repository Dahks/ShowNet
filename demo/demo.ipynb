```python
import igraph as ig

g = ig.Graph.Erdos_Renyi(n=15, m=30, directed=False, loops=False)
g.to_directed(mode="acyclic")
```

```python
"""
Old ugly default IGRAPH plotting with matplotlib
"""
import matplotlib.pyplot as plt 
fig, ax = plt.subplots()
ig.plot(
  g,
  target=ax,
  layout="sugiyama",
)
plt.show()
```
